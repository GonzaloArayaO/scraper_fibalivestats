#### **Descargar partidos de manera individual**

```python
# PARAMETROS A MODIFICAR

match_id = 2607028
fecha = '2024-12-15'
fase = 'Final four'
competencia = 'LNB 2'
export_csv = True
```

#### **Descargar partidos desde tabla_partidos**

```python
# Lectura tabla con datos de partidos
tabla_partidos = pd.read_excel(f'../data/tabla_partidos.xlsx')
```

```python
# Formato de datos
tabla_partidos['fecha'] = tabla_partidos['fecha'].dt.date       # fecha en formato correcto
```

```python
# Función para scrapear box_score
print('''
    Libreria scraper_fibalivestats
    - Función: box_score
    ------------------------------
    ''')
# Iterar sobre las filas del DataFrame
for index, row in tabla_partidos.iterrows():
    match_id = row['match_id']
    fecha = row['fecha']
    fase = row['fase']
    export_csv = True

    # Llamar a la función con los parámetros actuales
    box_score(match_id, fecha, fase, export_csv)
    time.sleep(np.random.randint(1, 5))
    print(f'Partido descargado: {match_id}')
```
